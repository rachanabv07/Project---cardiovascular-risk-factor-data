# Project : Cardiovascular Risk Factor Data

## About Dataset

- DataSet is taken from Kaggle - https://www.kaggle.com/datasets/mamta1999/cardiovascular-risk-data 
- DataSet provided information on over 4,000 patients and included 15 attributes, each representing a potential risk factor for CHD. These attributes included demographic, behavioral, and medical risk factors. 


## Variables Description

### Demographic:

- Sex: male or female ("M" or "F")
Age: Age of the patient (Continuous - Although the recorded ages have been truncated to whole numbers, the concept of age is continuous)
- Education: The level of education of the patient (categorical values - 1,2,3,4)

### Behavioral:

- is_smoking: whether or not the patient is a current smoker ("YES" or "NO")
- Cigs Per Day: the number of cigarettes that the person smoked on average in one day.(can be considered continuous as one can have any number of cigarettes, even half a cigarette.)

### Medical (history):

- BP Meds: whether or not the patient was on blood pressure medication (Nominal)
- Prevalent Stroke: whether or not the patient had previously had a stroke (Nominal)
- Prevalent Hyp: whether or not the patient was hypertensive (Nominal)
- Diabetes: whether or not the patient had diabetes (Nominal)

### Medical (current):

- Tot Chol: total cholesterol level (Continuous)
- Sys BP: systolic blood pressure (Continuous)
- Dia BP: diastolic blood pressure (Continuous)
- BMI: Body Mass Index (Continuous)
- Heart Rate: heart rate (Continuous - In medical research, variables such as heart rate though in fact discrete, yet are considered continuous because of large number of possible values.)
- Glucose: glucose level (Continuous)

### Predict variable (desired target):

10-year risk of coronary heart disease CHD(binary: “1”, means “Yes”, “0” means “No”)

# Models

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import classification_report
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
```

```python
model_params = {
    "SVM": {"model":svm.SVC(gamma = "auto", random_state=42, probability=True), "params":{"C": [0.01,1,10], "kernel":["rbf","linear"]}},
    "Decision_Tree":{"model":DecisionTreeClassifier(random_state=42), "params":{'criterion': ['gini', 'entropy'],'max_depth': [None, 2,5,10,20]}},
    "Random_Forest":{"model":RandomForestClassifier(random_state=42), "params":{"n_estimators":[1,500 ],"max_depth":[2,3,5]}},
    "Logistic_Regression":{"model": LogisticRegression(solver='liblinear',random_state=42),"params":{"C":[1,0.01,0.001]}}
}
```

```python
scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp["model"], mp["params"], cv = 10, return_train_score=False)
    clf.fit(train_set, y_train)
    scores.append({"model": model_name,
                   "best_score":clf.best_score_,
                  "best_params": clf.best_params_,
                  "best_estimator":clf.best_estimator_})
```

```python
df = pd.DataFrame(scores, columns = ["model", "best_score", "best_params","best_estimator"])
df
```

```python
def clas_report(model):
    y_pretrain = model.predict(train_set)
    print("."*10," classification_report on Train","."*10)
    print(classification_report(y_train, y_pretrain))
    print("."*10," classification_report on pre","."*10)
    print(classification_report(y_test, y_pre))
    
```

```python
def ROC_plot(model):
    proba = model.predict_proba(test_set)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, proba)
    roc_auc = auc(fpr, tpr)
    print("ROC-AUC = ", roc_auc) 
    
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b',label='AUC = %0.3f'% roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0,1],[0,1],'r--')
    plt.xlim([-0.1,1.0])
    plt.ylim([-0.1,1.01])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
```

```python
for models in range(len(df)):
    model = df["best_estimator"][models]
    model_name = df["model"][models]
    print("="*20,model_name,"="*20)
    print(model_name,"score=",model.score(test_set, y_test))
    y_pre = model.predict(test_set)
    y_pretrain = model.predict(train_set)
    proba = model.predict_proba(test_set)[:, 1]

    clas_report(model)
    ROC_plot(model)
    
```

## Ensemble Model

```python
LR=LogisticRegression(solver='liblinear', random_state=42)
DT = DecisionTreeClassifier(max_depth=3, random_state=42)
RF = RandomForestClassifier(random_state=42)
SVMc = SVC(probability=True, random_state=42)
```

```python
ensemble_model = VotingClassifier(estimators=[('LR',LR),('RF', RF), ('DT', DT), ('SVMc', SVMc)], voting='soft')
ensemble_model.fit(train_set, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_predict = ensemble_model.predict(test_set)
accuracy = accuracy_score(y_predict, y_test)
print('Accuracy of the Ensemble Model on the test set:', accuracy)

clas_report(ensemble_model)
ROC_plot(ensemble_model)
```
